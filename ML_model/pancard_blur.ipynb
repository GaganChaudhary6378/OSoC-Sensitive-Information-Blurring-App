```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


image_path =  '/content/test.jpg'  # Load the PAN card image
image = cv2.imread(image_path)


photo_coordinates = [(100, 100), (280, 300)]
name_coordinates = [(150, 400), (550, 450)]
dob_coordinates = [(150, 500), (300, 550)]
card_no_coordinates = [(100, 100), (400, 150)]

photo_coordinates = [
    (photo_coordinates[0][0] - 70, photo_coordinates[0][1]+30),
    (photo_coordinates[1][0] - 70, photo_coordinates[1][1]+30)
]

y_name_coordinates = [
    (name_coordinates[0][0] - 150, name_coordinates[0][1]-40),
    (name_coordinates[1][0] - 150, name_coordinates[1][1]-40)
]
name_coordinates = [
    (name_coordinates[0][0] - 150, name_coordinates[0][1]+38),
    (name_coordinates[1][0] - 150, name_coordinates[1][1]+38)
]
dob_coordinates = [
    (dob_coordinates[0][0] - 150, dob_coordinates[0][1]+50),
    (dob_coordinates[1][0] - 150, dob_coordinates[1][1]+50)
]
pan_coordinates = [
    (card_no_coordinates[0][0]  +170, card_no_coordinates[0][1] +140),
    (card_no_coordinates[1][0]  +170, card_no_coordinates[1][1] +140)
]



blurred_image = np.copy(image)
kernel_size=(55,55)

blurred_image[photo_coordinates[0][1]:photo_coordinates[1][1], photo_coordinates[0][0]:photo_coordinates[1][0]] = cv2.GaussianBlur(
    image[photo_coordinates[0][1]:photo_coordinates[1][1], photo_coordinates[0][0]:photo_coordinates[1][0]], kernel_size,0
)


blurred_image[y_name_coordinates[0][1]:y_name_coordinates[1][1], y_name_coordinates[0][0]:y_name_coordinates[1][0]] = cv2.GaussianBlur(
    image[y_name_coordinates[0][1]:y_name_coordinates[1][1], y_name_coordinates[0][0]:y_name_coordinates[1][0]], kernel_size ,0
)


blurred_image[name_coordinates[0][1]:name_coordinates[1][1], name_coordinates[0][0]:name_coordinates[1][0]] = cv2.GaussianBlur(
    image[name_coordinates[0][1]:name_coordinates[1][1], name_coordinates[0][0]:name_coordinates[1][0]], kernel_size ,0
)


blurred_image[dob_coordinates[0][1]:dob_coordinates[1][1], dob_coordinates[0][0]:dob_coordinates[1][0]] = cv2.GaussianBlur(
    image[dob_coordinates[0][1]:dob_coordinates[1][1], dob_coordinates[0][0]:dob_coordinates[1][0]], kernel_size ,0
)

blurred_image[pan_coordinates[0][1]:pan_coordinates[1][1], pan_coordinates[0][0]:pan_coordinates[1][0]] = cv2.GaussianBlur(
    image[pan_coordinates[0][1]:pan_coordinates[1][1], pan_coordinates[0][0]:pan_coordinates[1][0]], kernel_size ,0
)

cv2_imshow(blurred_image)
```
